### Sometimes data may not be clear when plotted  making it challenging to interpret.Therefore, we should consider data rescaling to improve readability.

```python
x = [1, 2, 3, 4, 5]
y = [85, 92, 78, 88, 95]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

 ----

```python
# Calculate the range for x 
minx = min(x)
maxx = max(x)
range_column_x = maxx - minx
```

```python
# Calculate the range for y
miny = min(y)
maxy = max(y)
range_column_y = maxy - miny
```

```python
# Calculate the mean for x and y
meanx = np.mean(x)
meany = np.mean(y)
```

# xi=xi-Î¼/ Range xi
    This law allows you to rescale values between -1 and 1, with zero in     between, to facilitate data interpretation.

```python
# Define a function for rescaling
def rescaling(x, mean, range_column):
    rescaled_x = [(xi - mean) / range_column for xi in x]
    return rescaled_x
```

```python
# Rescale the x values
rescaled_x = rescaling(x, meanx, range_column_x)
# Rescale the y values (if needed)
rescaled_y = rescaling(y, meany, range_column_y)

```

```python
rescaled_y
```

```python
plt.plot(rescaled_x, rescaled_y)
plt.xlabel('X-number-room')
plt.ylabel('Y- price')
plt.grid(True)
plt.show()
```
